```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
vec1 = [1, 3, 5, 10, 14, 15, 16, 17, 20, 25, 35, 45, 60, 100]
vec2 = [5.08, 4.98, 4.88, 4.20, 3.76, 3.56, 3.40, 3.36, 3.04, 2.68, 2.08, 1.72, 1.32, 0.9]

# Convert lists to numpy arrays
vec1 = np.array(vec1)
vec2 = np.array(vec2)

# Calculate dB values
vec2_db = 20 * np.log10(vec2 / 5)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(vec1, vec2_db, marker='o', linestyle='-', color='b', label='Magnitude Response')

# Highlight the -3dB point
plt.axhline(y=-3, color='r', linestyle='--', label='-3dB')

# Scale the x-axis for better low-pass representation
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('Frequency Response of Low-Pass Filter')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```
